## **1️⃣ Introduction: ทำไมต้องใช้ Neural Networks?**  
ในปัจจุบัน **Artificial Neural Networks (ANNs)** เป็นหนึ่งในโมเดล Machine Learning ที่ทรงพลังที่สุด โดย ANN สามารถแก้ปัญหาที่โมเดล Machine Learning แบบดั้งเดิมทำได้ยาก เช่น การจดจำภาพ การวิเคราะห์ข้อความ และการพยากรณ์ข้อมูลในอนาคต  

**แต่ทำไมเราต้องใช้ ANN แทนที่โมเดลอื่น?**  
- **Non-Linearity**: ANN สามารถเรียนรู้ฟังก์ชันที่ซับซ้อนได้ดีกว่า Linear Models  
- **Feature Learning**: ไม่ต้องกำหนด Feature ด้วยมือแบบ Logistic Regression หรือ Decision Trees  
- **Generalization**: สามารถปรับให้ทำงานได้กับปัญหาหลากหลายประเภท

## **2️⃣ โครงสร้างพื้นฐานของ Neuron ใน Neural Networks**  
โมเดลของ ANN ได้รับแรงบันดาลใจจากโครงสร้างของสมองมนุษย์ ประกอบด้วย **Neuron** (เซลล์ประสาท) หลายตัวที่ทำงานร่วมกัน  

### **2.1 โครงสร้างของ Neuron**
Neuron แต่ละตัวประกอบด้วย:  
1. **Input** (ค่าเข้า) → ข้อมูลที่ใช้เป็นอินพุต เช่น รูปภาพ, ข้อความ, ตัวเลข  
2. **Weight (w)** → ค่าถ่วงน้ำหนักที่ปรับเพื่อเรียนรู้รูปแบบของข้อมูล  
3. **Bias (b)** → ค่าคงที่ที่ช่วยปรับค่าอินพุต  
4. **Activation Function (f)** → ฟังก์ชันที่ใช้กำหนดว่าข้อมูลจะส่งผ่านไปยังชั้นถัดไปหรือไม่  
5. **Output** (ค่าผลลัพธ์) → ผลที่ได้จากการประมวลผล  

**สมการพื้นฐานของ Neuron**  
\[
y = f\left(\sum_{i=1}^{n} w_i x_i + b\right)
\]  
โดยที่  
- \( x_i \) = ค่าอินพุตที่ \( i \)  
- \( w_i \) = ค่าถ่วงน้ำหนัก  
- \( b \) = bias  
- \( f \) = Activation Function
