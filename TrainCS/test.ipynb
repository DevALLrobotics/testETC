‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡∏™‡∏°‡∏°‡∏ï‡∏¥‡∏ß‡πà‡∏≤‡∏°‡∏µ‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç 4 ‡∏´‡∏•‡∏±‡∏Å ‡πÅ‡∏•‡∏∞‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡πÉ‡∏ä‡πâ brute force ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏î‡∏≤‡∏£‡∏´‡∏±‡∏™‡πÉ‡∏´‡πâ‡∏Ñ‡∏£‡∏ö‡∏ó‡∏∏‡∏Å‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏õ‡πá‡∏ô‡πÑ‡∏õ‡πÑ‡∏î‡πâ

```python
# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á Brute Force ‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô 4 ‡∏´‡∏•‡∏±‡∏Å
correct_password = "5271"  # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏£‡∏´‡∏±‡∏™‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á

for i in range(10000):  # 0000 ‡∏ñ‡∏∂‡∏á 9999
    attempt = str(i).zfill(4)  # ‡πÄ‡∏ï‡∏¥‡∏° 0 ‡∏î‡πâ‡∏≤‡∏ô‡∏´‡∏ô‡πâ‡∏≤‡πÉ‡∏´‡πâ‡∏Ñ‡∏£‡∏ö 4 ‡∏´‡∏•‡∏±‡∏Å
    print(f"Trying: {attempt}")
    if attempt == correct_password:
        print(f"Password found: {attempt}")
        break
```

üî• ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á 2: Brute Force ‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£

‡∏ñ‡πâ‡∏≤‡πÄ‡∏£‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£ brute force ‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£ ‡πÄ‡∏ä‡πà‡∏ô "abc" ‡∏´‡∏£‡∏∑‡∏≠ "xyz" ‡πÄ‡∏£‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡πÉ‡∏ä‡πâ

```python
import itertools
import string

password = "abc"  # ‡∏£‡∏´‡∏±‡∏™‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á
chars = string.ascii_lowercase  # 'abcdefghijklmnopqrstuvwxyz'

for attempt in itertools.product(chars, repeat=len(password)):
    attempt = ''.join(attempt)
    print(f"Trying: {attempt}")
    if attempt == password:
        print(f"Password found: {attempt}")
        break
```

üî• ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á 3: Brute Force ‡πÉ‡∏ä‡πâ Multithreading ‡πÉ‡∏´‡πâ‡πÄ‡∏£‡πá‡∏ß‡∏Ç‡∏∂‡πâ‡∏ô

Brute Force ‡∏ò‡∏£‡∏£‡∏°‡∏î‡∏≤‡∏ä‡πâ‡∏≤‡πÉ‡∏ä‡πà‡πÑ‡∏´‡∏°? ‡∏•‡∏≠‡∏á‡πÉ‡∏ä‡πâ Multithreading ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏£‡πá‡∏ß

```python
import itertools
import string
import threading

password = "xyz"  # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏£‡∏´‡∏±‡∏™‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á
chars = string.ascii_lowercase
found = False  # ‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡∏ö‡∏≠‡∏Å‡∏ß‡πà‡∏≤‡∏´‡∏≤‡πÄ‡∏à‡∏≠‡∏´‡∏£‡∏∑‡∏≠‡∏¢‡∏±‡∏á

def brute_force(start, end):
    global found
    for attempt in itertools.islice(itertools.product(chars, repeat=len(password)), start, end):
        if found:
            return
        attempt = ''.join(attempt)
        print(f"Trying: {attempt}")
        if attempt == password:
            print(f"Password found: {attempt}")
            found = True
            return

# ‡πÅ‡∏ö‡πà‡∏á‡πÄ‡∏õ‡πá‡∏ô 4 Threads
threads = []
chunk_size = len(chars) ** len(password) // 4
for i in range(4):
    start = i * chunk_size
    end = (i + 1) * chunk_size if i != 3 else None  # Thread ‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢‡πÉ‡∏´‡πâ‡∏£‡∏±‡∏ô‡∏ñ‡∏∂‡∏á‡∏à‡∏ö
    t = threading.Thread(target=brute_force, args=(start, end))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```
